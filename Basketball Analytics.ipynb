Filtering in data frames

```python
Filter=stats2.InternetUsers<2
stats2[Filter]
```

```python
Filter2=stats2.BirthRate>40
stats2[Filter2]      
#alt    stats2[stats2.BirthRate>40]
```

```python
## stats[Filter and Filter2] gives error - Not a bitwise operator
## and -> expects all rows in filter1 and filter2 to be either true or false
##-- Hence, we will use bitwise and &
stats2[Filter & Filter2]
```

```python
## stats2[stats2.BirthRate>40&stats2.InternetUsers<2] -> Error : Python does not know which logical block to solve first
stats2[(stats2.BirthRate>40)&(stats2.InternetUsers<2)]
```

Accessing individual elements
iat - looks for integer, actually counts the index
at - looks for label, takes the index from the label

```python
stats2.head()
```

```python
stats2.iat[3,4]
```

```python
stats2.iat[0,1]
```

```python
stats2.at[2,'BirthRate']
```

```python
###---- Differences between at and iat
sub10=stats2[::10]
```

```python
sub10
```

```python
###--- iat looks for integer - 10th row in 0th col
sub10.iat[10,0]
```

```python
###--- at looks for label - fetches value where index is 10
sub10.at[10,'CountryName']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams['figure.figsize']=8,4
```

```python
###---- Distribution
viz1=sns.distplot(stats2['InternetUsers'])
```

```python
#creating bins in seaborn
viz1=sns.distplot(stats2['InternetUsers'], bins=30, )
```

```python
#creating a boxplot
viz2=sns.boxplot(data=stats2, x='IncomeGroup', y='BirthRate')
```

visualizing using seaborn

```python
#Linear regression model
viz3=sns.lmplot(data=stats2, x='InternetUsers', y='BirthRate', fit_reg=False, size=10)
```

```python
viz4=sns.lmplot(data=stats2, x='InternetUsers', y='BirthRate', fit_reg=False, size=10, hue='IncomeGroup')
```
